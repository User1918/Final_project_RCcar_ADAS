```python
# --- Dataset Paths ---

# 1. Path to the EXISTING dataset FOLDER TO BE MODIFIED (Originally 17 classes)
target_dataset_base_dir = '/content/drive/MyDrive/Combine_dataset' # <--- EDIT THIS (Your 'old' dataset path)

# 2. Path to the NEW 1-class dataset (from Roboflow) to be added
new_dataset_base_dir = '/content/drive/MyDrive/Car_dataset'  # <--- EDIT THIS

# --- Class ID ---
# Assign a unique ID to the class from the new dataset
new_class_id = 18

# --- Dataset Structure Settings ---
# Does your TARGET (Train_yolo) dataset use 'images' and 'labels' subfolders?
use_images_labels_subfolders_TARGET = True # <--- EDIT THIS (True or False)

# Does your NEW (Roboflow) dataset use 'images' and 'labels' subfolders?
use_images_labels_subfolders_NEW = True # <--- EDIT THIS (True or False)

# --- Filename Conflict Handling ---
# If True, rename files from the new dataset (prefix 'new_class_') if a file
# with the same name already exists in the target dataset. Prevents overwriting.
# If False, files with the same name WILL BE OVERWRITTEN (Use with caution!).
rename_new_files_on_conflict = True # Set to False ONLY if you understand the risk

# --- End Configuration ---

# --- Print configuration for verification ---
print("--- Configuration Summary (In-Place Merge) ---")
print(f"Target (Existing) Dataset Path: {target_dataset_base_dir} <--- WILL BE MODIFIED")
print(f"  Structure uses images/labels subfolders: {use_images_labels_subfolders_TARGET}")
print(f"New Dataset Path: {new_dataset_base_dir}")
print(f"  Structure uses images/labels subfolders: {use_images_labels_subfolders_NEW}")
print(f"New Class ID to Add: {new_class_id}")
print(f"Rename new files on conflict: {rename_new_files_on_conflict}")
print("-----------------------------")
print(" WARNING: The TARGET dataset folder listed above will be directly modified. ")
print(" It is strongly recommended to BACK UP the target folder before proceeding. ")
print("-----------------------------")


# --- Basic Path Validation ---
if not os.path.isdir(target_dataset_base_dir):
    print(f"!!! CRITICAL ERROR: Target dataset directory NOT FOUND: {target_dataset_base_dir}")
if not os.path.isdir(new_dataset_base_dir):
    print(f"!!! WARNING: New dataset directory NOT FOUND: {new_dataset_base_dir}")

print("\nPlease carefully review the paths, settings, and warnings above.")
```

```python
print("--- Starting: Processing and ADDING NEW Dataset Files to Target Dataset ---")

# Ensure target directory exists from configuration check
if not os.path.isdir(target_dataset_base_dir):
     print(f"FATAL ERROR: Target directory {target_dataset_base_dir} not found. Aborting.")
     # Raise an error or exit if you prefer stronger error handling
else:
    for split in ['train', 'valid', 'test']:
        print(f"\nProcessing NEW dataset split: {split} -> adding to TARGET '{split}'")
        new_split_dir = os.path.join(new_dataset_base_dir, split)

        # Determine NEW dataset source paths
        new_img_src_dir = os.path.join(new_split_dir, 'images') if use_images_labels_subfolders_NEW else new_split_dir
        new_lbl_src_dir = os.path.join(new_split_dir, 'labels') if use_images_labels_subfolders_NEW else new_split_dir

        # Determine TARGET dataset destination paths (where files will be added/modified)
        target_img_dest_dir = os.path.join(target_dataset_base_dir, split, 'images') if use_images_labels_subfolders_TARGET else os.path.join(target_dataset_base_dir, split)
        target_lbl_dest_dir = os.path.join(target_dataset_base_dir, split, 'labels') if use_images_labels_subfolders_TARGET else os.path.join(target_dataset_base_dir, split)

        processed_img_count = 0
        processed_lbl_count = 0
        skipped_conflict_count = 0
        overwritten_count = 0
        error_count = 0

        # Ensure target destination directories exist (they should from the original dataset)
        if not os.path.isdir(target_img_dest_dir):
             print(f"  Warning: TARGET image destination directory not found, skipping images for this split: {target_img_dest_dir}")
             # Decide whether to create it or skip
             # os.makedirs(target_img_dest_dir, exist_ok=True) # Uncomment to create if missing
             # continue # Uncomment to skip split if image folder structure missing in target
        if not os.path.isdir(target_lbl_dest_dir):
             print(f"  Warning: TARGET label destination directory not found, skipping labels for this split: {target_lbl_dest_dir}")
             # os.makedirs(target_lbl_dest_dir, exist_ok=True) # Uncomment to create if missing
             # continue # Uncomment to skip split if label folder structure missing in target

        # Check if source directories exist for the new dataset's split
        if not os.path.isdir(new_lbl_src_dir):
            print(f"  Info: NEW label source directory not found for split '{split}', nothing to add from here: {new_lbl_src_dir}")
            continue # Skip this split if no source labels folder
        if not os.path.isdir(new_img_src_dir):
            print(f"  Warning: NEW image source directory not found for split '{split}', skipping image copying: {new_img_src_dir}")
            # This allows processing labels even if images are missing, but it's unusual.

        print(f"  Processing labels from '{new_lbl_src_dir}'...")
        # Iterate through label files first
        for lbl_filename in os.listdir(new_lbl_src_dir):
            if not lbl_filename.lower().endswith('.txt'):
                continue

            base_name, _ = os.path.splitext(lbl_filename)
            lbl_source_path = os.path.join(new_lbl_src_dir, lbl_filename)

            # Find corresponding image file in the source
            img_filename = None
            img_source_path = None
            if os.path.isdir(new_img_src_dir): # Only look for images if src dir exists
                for ext in ['.png', '.jpg', '.jpeg']:
                    potential_img_path = os.path.join(new_img_src_dir, base_name + ext)
                    if os.path.exists(potential_img_path):
                        img_filename = base_name + ext
                        img_source_path = potential_img_path
                        break

            if not img_source_path and os.path.isdir(new_img_src_dir):
                print(f"    Warning: No corresponding image found for label '{lbl_filename}' in source. Skipping label processing for this file.")
                error_count += 1
                continue
            elif not os.path.isdir(new_img_src_dir):
                 print(f"    Info: Processing label '{lbl_filename}' without copying image (source image dir missing).")
                 img_filename = None # Ensure we don't try to copy image later

            # --- Determine Destination Filenames & Handle Conflicts ---
            dest_lbl_filename = lbl_filename
            dest_img_filename = img_filename # Can be None if img_source_path is None

            lbl_dest_path = os.path.join(target_lbl_dest_dir, dest_lbl_filename)
            img_dest_path = os.path.join(target_img_dest_dir, dest_img_filename) if dest_img_filename else None

            # Check for conflicts
            conflict_exists = os.path.exists(lbl_dest_path) or (img_dest_path and os.path.exists(img_dest_path))

            if conflict_exists:
                if rename_new_files_on_conflict:
                    print(f"    Info: Conflict for '{lbl_filename}'/'{img_filename}'. Renaming with 'new_class_' prefix.")
                    original_lbl_filename = dest_lbl_filename
                    original_img_filename = dest_img_filename
                    dest_lbl_filename = f"new_class_{original_lbl_filename}"
                    dest_img_filename = f"new_class_{original_img_filename}" if original_img_filename else None

                    lbl_dest_path = os.path.join(target_lbl_dest_dir, dest_lbl_filename)
                    img_dest_path = os.path.join(target_img_dest_dir, dest_img_filename) if dest_img_filename else None

                    # Check again if even the renamed file exists
                    if os.path.exists(lbl_dest_path) or (img_dest_path and os.path.exists(img_dest_path)):
                        print(f"    Warning: Renamed file '{dest_lbl_filename}'/'{dest_img_filename}' also exists. Skipping.")
                        skipped_conflict_count += 1
                        continue
                else:
                    # rename_new_files_on_conflict is False, so we will overwrite
                    print(f"    Warning: Overwriting existing file(s) for '{lbl_filename}'/'{img_filename}' (rename_on_conflict=False).")
                    overwritten_count += 1 # Technically overwriting label and maybe image

            # --- Process Label File (Change Class ID) ---
            try:
                modified_lines = []
                with open(lbl_source_path, 'r') as infile:
                    for line_num, line in enumerate(infile):
                        parts = line.strip().split()
                        if len(parts) >= 5:
                            try:
                                original_class_id = int(parts[0])
                                if original_class_id == 0:
                                    parts[0] = str(new_class_id) # Replace with 17
                                    modified_lines.append(" ".join(parts))
                                else:
                                    print(f"      Warning: Found unexpected class ID {original_class_id} in new dataset file '{lbl_filename}', line {line_num+1}. Keeping original.")
                                    modified_lines.append(line.strip())
                            except ValueError:
                                 print(f"      Warning: Non-integer class ID in '{lbl_filename}', line {line_num+1}. Skipping line.")
                        elif line.strip():
                             print(f"      Warning: Malformed line in '{lbl_filename}', line {line_num+1}. Skipping line.")

                # Write modified label file to TARGET destination
                with open(lbl_dest_path, 'w') as outfile:
                    for line in modified_lines:
                        outfile.write(line + "\n")
                processed_lbl_count += 1

                # Copy corresponding image file to TARGET destination (if it exists and path is set)
                if img_source_path and img_dest_path:
                    shutil.copy2(img_source_path, img_dest_path)
                    processed_img_count += 1
                elif not img_dest_path and img_filename: # Implies source image dir was missing
                    pass # Label processed, image skipped as noted earlier
                # else: # img_filename was None or img_dest_path was None


            except Exception as e:
                print(f"    ERROR processing file pair '{lbl_filename}'/'{img_filename}': {e}")
                error_count += 1

        print(f"  Split '{split}' summary:")
        print(f"    Labels Processed/Written to Target: {processed_lbl_count}")
        print(f"    Images Copied to Target: {processed_img_count}")
        print(f"    Skipped (due to conflicts/existing): {skipped_conflict_count}")
        print(f"    Overwritten (rename_on_conflict=False): {overwritten_count}")
        print(f"    Errors/Warnings: {error_count}")

    print("\n--- Finished: Processing and ADDING NEW Dataset Files to Target Dataset ---")
```

```python
print("--- Verification: File Counts in MODIFIED Target Dataset ---")
print(f"Checking folder: {target_dataset_base_dir}")

total_images = 0
total_labels = 0

for split in ['train', 'valid', 'test']:
    print(f"\nVerifying split: {split}")
    split_dir = os.path.join(target_dataset_base_dir, split)

    # Determine target paths based on target structure setting
    img_folder = os.path.join(split_dir, 'images') if use_images_labels_subfolders_TARGET else split_dir
    lbl_folder = os.path.join(split_dir, 'labels') if use_images_labels_subfolders_TARGET else split_dir

    # Count image files
    img_count = 0
    if os.path.isdir(img_folder):
        img_files = glob.glob(os.path.join(img_folder, '*.jpg')) + \
                    glob.glob(os.path.join(img_folder, '*.jpeg')) + \
                    glob.glob(os.path.join(img_folder, '*.png'))
        img_count = len(img_files)
        print(f"  Image files (.jpg, .jpeg, .png) found in '{img_folder}': {img_count}")
        total_images += img_count
    else:
        print(f"  Image folder not found: {img_folder}")


    # Count label files
    lbl_count = 0
    if os.path.isdir(lbl_folder):
        lbl_files = glob.glob(os.path.join(lbl_folder, '*.txt'))
        lbl_count = len(lbl_files)
        print(f"  Label files (.txt) found in '{lbl_folder}': {lbl_count}")
        total_labels += lbl_count
    else:
        print(f"  Label folder not found: {lbl_folder}")

    if img_count != lbl_count:
         print(f"  Warning: Mismatch between image count ({img_count}) and label count ({lbl_count}) in split '{split}'.")

print("\n-----------------------------")
print(f"Total Target Images Now: {total_images}")
print(f"Total Target Labels Now: {total_labels}")
if total_images != total_labels:
    print("Warning: Overall image and label counts do not match!")
print("-----------------------------")
print("Note: Counts should now be higher than the original dataset.")
```

```python
print("--- Verification: Class IDs in MODIFIED Target Dataset Labels ---")
target_dataset_base_dir = '/content/drive/MyDrive/Combine_dataset'
use_images_labels_subfolders_TARGET = True # <--- EDIT THIS (True or False)
expected_min_id = 0
expected_max_id = 20
expected_total_classes = 21

all_splits_ok = True

for split in ['train', 'valid', 'test']:
    print(f"\nVerifying split: {split}")
    split_dir = os.path.join(target_dataset_base_dir, split)

    # Determine target label path
    lbl_folder = os.path.join(split_dir, 'labels') if use_images_labels_subfolders_TARGET else split_dir

    if not os.path.isdir(lbl_folder):
        print(f"  Label folder not found, skipping check: {lbl_folder}")
        # Consider if this split *should* have labels. If so, it's an issue.
        # all_splits_ok = False # Uncomment if missing label folder is an error condition
        continue

    unique_class_ids = set()
    processed_file_count = 0
    total_txt_files = 0
    files_in_dir = os.listdir(lbl_folder)

    for filename in files_in_dir:
        if filename.lower().endswith('.txt'):
            total_txt_files += 1
            file_path = os.path.join(lbl_folder, filename)
            try:
                with open(file_path, 'r') as f:
                    for line_num, line in enumerate(f):
                        parts = line.strip().split()
                        if len(parts) > 0:
                            try:
                                class_id = int(parts[0])
                                unique_class_ids.add(class_id)
                            except ValueError:
                                print(f"    Warning: Non-integer ID in '{filename}', line {line_num + 1}: '{line.strip()}'")
                processed_file_count += 1
            except Exception as e:
                print(f"    Error reading file '{filename}': {e}")

    print(f"  Scan complete. Processed {processed_file_count} out of {total_txt_files} '.txt' files found.")

    if not unique_class_ids:
        print("  No valid class IDs found in this split.")
        if total_txt_files > 0 : all_splits_ok = False # If txt files exist but no IDs found, it's likely an issue
    else:
        sorted_ids = sorted(list(unique_class_ids))
        min_id = min(sorted_ids)
        max_id = max(sorted_ids)
        num_classes = len(sorted_ids)

        print(f"  Unique Class IDs found: {sorted_ids}")
        print(f"  Min ID: {min_id}, Max ID: {max_id}, Count: {num_classes}")

        # Check against expectations
        split_ok = True
        if min_id != expected_min_id:
            print(f"    Warning: Minimum ID is {min_id}, expected {expected_min_id}.")
            split_ok = False
        # Check if the max ID is at least the new ID (it could be exactly new_class_id or higher if original had more)
        if max_id < new_class_id:
             print(f"    Warning: Maximum ID is {max_id}, but expected at least {new_class_id} after adding the new class.")
             split_ok = False
        elif max_id > new_class_id:
             print(f"    Info: Maximum ID {max_id} is higher than the newly added ID {new_class_id}. This is OK if your original dataset already had IDs up to {max_id}.")
        # Check if the new class ID is present
        if new_class_id not in unique_class_ids:
            print(f"    Warning: The new class ID {new_class_id} was NOT found in this split's labels.")
            split_ok = False


        if split_ok and max_id == new_class_id and min_id == expected_min_id and num_classes == expected_total_classes:
             print(f"    Class ID range (0-{new_class_id}) and count ({expected_total_classes}) appear correct for this split.")
        elif split_ok and new_class_id in unique_class_ids:
             print(f"    Class ID range seems OK, and new class ID {new_class_id} is present.")
        else:
             all_splits_ok = False # Mark overall as not OK if specific checks fail

print("\n--- Overall Class ID Check Summary ---")
if all_splits_ok:
    print("All splits seem to have the correct minimum ID (0) and contain the new class ID (17).")
    print(f"Check Max ID and Count warnings above based on your original dataset's range (expected max {new_class_id}).")
else:
    print("One or more splits have unexpected class IDs or are missing the new class ID. Please review warnings above.")
print("------------------------------------")
```

```python
import os
from pathlib import Path
import logging

# --- Configuration ---
# Set the base path to your dataset on Google Drive
dataset_base_path = Path("/content/drive/MyDrive/Combine_dataset")

# Define the splits to check (e.g., train, valid, test)
splits_to_check = ["train", "valid","test"]

# Define common image extensions to look for
image_extensions = [".jpg", ".jpeg", ".png", ".bmp", ".webp"]

# Setup logging
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')
# --- End Configuration ---

def check_dataset_integrity(base_path: Path, splits: list[str], img_exts: list[str]):
    """
    Checks for inconsistencies between image and label files in specified dataset splits.

    Args:
        base_path: The root directory of the dataset.
        splits: A list of dataset splits (subdirectories) to check (e.g., ['train', 'valid']).
        img_exts: A list of valid image file extensions (e.g., ['.jpg', '.png']).
    """
    all_ok = True

    for split in splits:
        logging.info(f"--- Checking Split: '{split}' ---")
        image_dir = base_path / split / "images"
        label_dir = base_path / split / "labels"

        if not image_dir.is_dir():
            logging.warning(f"Image directory not found for split '{split}': {image_dir}")
            continue
        if not label_dir.is_dir():
            logging.warning(f"Label directory not found for split '{split}': {label_dir}")
            continue

        # Get sets of filenames (without extensions)
        try:
            image_files = {p.stem for p in image_dir.iterdir() if p.suffix.lower() in img_exts and p.is_file()}
            label_files = {p.stem for p in label_dir.iterdir() if p.suffix.lower() == ".txt" and p.is_file()}
        except Exception as e:
            logging.error(f"Error reading files in split '{split}': {e}. Check permissions or path.")
            all_ok = False
            continue


        # Check 1: Labels without corresponding images
        labels_missing_images = label_files - image_files
        if labels_missing_images:
            all_ok = False
            logging.warning(f"Found {len(labels_missing_images)} label files in '{split}' missing corresponding images:")
            for filename_stem in sorted(list(labels_missing_images)):
                print(f"  - Label: {label_dir / (filename_stem + '.txt')}")
            print(f"  RECOMMENDATION: Delete these orphaned label files from {label_dir}")
        else:
            logging.info(f"All label files in '{split}' have corresponding images. ")

        # Check 2: Images without corresponding labels
        images_missing_labels = image_files - label_files
        if images_missing_labels:
            # Note: This might be acceptable if you have background images without objects
            logging.warning(f"Found {len(images_missing_labels)} image files in '{split}' missing corresponding label files:")
            for filename_stem in sorted(list(images_missing_labels)):
                # Find the actual image file extension
                actual_image_file = next((p for p in image_dir.iterdir() if p.stem == filename_stem and p.suffix.lower() in img_exts), None)
                if actual_image_file:
                     print(f"  - Image: {actual_image_file}")
                else:
                     print(f"  - Image stem (extension unknown): {image_dir / filename_stem}") # Fallback
            print(f"  RECOMMENDATION: If these are not background images, create label files or remove the images from {image_dir}")

        else:
            logging.info(f"All image files in '{split}' have corresponding label files. ")

        print("-" * (len(split) + 20)) # Separator

    if all_ok:
        logging.info("=== Dataset Integrity Check Complete: No major inconsistencies found. ===")
    else:
        logging.warning("=== Dataset Integrity Check Complete: Inconsistencies found. Please review the recommendations above. ===")

# --- Run the Check ---
if __name__ == "__main__":
    if not dataset_base_path.is_dir():
         logging.error(f"Dataset base path not found: {dataset_base_path}")
         logging.error("Please ensure the 'dataset_base_path' variable is set correctly and your Google Drive is mounted.")
    else:
        check_dataset_integrity(dataset_base_path, splits_to_check, image_extensions)
```

```python
# -*- coding: utf-8 -*-
# This cell writes the conversion script to a file named 'convert_labels.py'

import os
from pathlib import Path
import logging
import argparse # For command-line arguments

# --- Script Content ---
# Defines the Python code that will be written to the file
script_content = """
# -*- coding: utf-8 -*-
import os
from pathlib import Path
import logging
import argparse # For command-line arguments

# Setup logging
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')

def convert_file(input_path: Path, output_path: Path):
    \"\"\"
    Converts a single segmentation label file to a detection label file.

    Args:
        input_path: Path to the input segmentation label file (.txt).
        output_path: Path where the output detection label file (.txt) will be saved.
    \"\"\"
    try:
        output_lines = []
        # Open and read the input segmentation file line by line
        with open(input_path, 'r') as infile:
            for line_num, line in enumerate(infile, 1):
                parts = line.strip().split() # Split line into parts
                if not parts:
                    continue # Skip empty lines

                # Check for minimum required parts (class_id + one point) and odd number of coordinate values
                if len(parts) < 3 or len(parts) % 2 == 0:
                    logging.warning(f"Skipping malformed line {line_num} in {input_path}: Incorrect number of parts ({len(parts)}). Line: '{line.strip()}'")
                    continue

                try:
                    # Extract class ID and convert coordinates to floats
                    class_id = int(parts[0])
                    coords = [float(p) for p in parts[1:]]
                except ValueError:
                    # Handle cases where coordinate values are not numbers
                    logging.warning(f"Skipping malformed line {line_num} in {input_path}: Non-numeric value found. Line: '{line.strip()}'")
                    continue

                # Separate x and y coordinates from the list
                x_coords = coords[0::2] # Get elements at even indices (0, 2, 4, ...)
                y_coords = coords[1::2] # Get elements at odd indices (1, 3, 5, ...)

                # Ensure coordinates were actually extracted
                if not x_coords or not y_coords:
                     logging.warning(f"Skipping line {line_num} in {input_path}: No valid coordinates found after parsing. Line: '{line.strip()}'")
                     continue

                # Find the min/max x and y values to define the bounding box
                min_x = min(x_coords)
                max_x = max(x_coords)
                min_y = min(y_coords)
                max_y = max(y_coords)

                # Calculate YOLO detection format parameters (normalized)
                center_x = (min_x + max_x) / 2.0
                center_y = (min_y + max_y) / 2.0
                width = max_x - min_x
                height = max_y - min_y

                # Validate calculated bounding box values
                if not (0 <= center_x <= 1 and 0 <= center_y <= 1 and width >= 0 and height >= 0):
                     logging.warning(f"Skipping line {line_num} in {input_path}: Calculated bbox invalid or outside bounds (center=[{center_x:.4f},{center_y:.4f}], w={width:.4f}, h={height:.4f}). Original line: '{line.strip()}'")
                     continue
                # Log if box has zero width or height (might be intended or an issue)
                if width == 0 or height == 0:
                    logging.warning(f"Line {line_num} in {input_path} resulted in zero width or height (w={width:.4f}, h={height:.4f}). Keeping it, but check if this is intended. Original line: '{line.strip()}'")

                # Format the output line string
                output_line = f"{class_id} {center_x:.6f} {center_y:.6f} {width:.6f} {height:.6f}"
                output_lines.append(output_line) # Add to list of lines for this file

        # Write the converted detection labels to the output file
        output_path.parent.mkdir(parents=True, exist_ok=True) # Ensure output directory exists
        if output_lines: # Only create/write file if there are valid lines
            with open(output_path, 'w') as outfile:
                outfile.write("\\n".join(output_lines) + "\\n") # Join lines with newline and add trailing newline
        elif not output_path.exists():
             # Create an empty file if input was empty/invalid, prevents missing label errors later
             output_path.touch()

    except FileNotFoundError:
        logging.error(f"Input file not found: {input_path}")
    except Exception as e:
        logging.error(f"An unexpected error occurred processing {input_path}: {e}")

def convert_directory(input_dir: Path, output_dir: Path):
    \"\"\"
    Converts all .txt files in an input directory from segmentation to detection format.

    Args:
        input_dir: Directory containing the input segmentation label files (.txt).
        output_dir: Directory where the output detection label files (.txt) will be saved.
    \"\"\"
    if not input_dir.is_dir():
        logging.error(f"Input directory not found or is not a directory: {input_dir}")
        return

    output_dir.mkdir(parents=True, exist_ok=True) # Create output dir if needed

    converted_count = 0
    skipped_count = 0
    processed_files = set() # To avoid reprocessing if run multiple times

    logging.info(f"Scanning input directory: {input_dir}")
    try:
        items_in_dir = list(input_dir.iterdir()) # Get directory contents
        logging.info(f"Found {len(items_in_dir)} items in input directory.")
    except Exception as e:
        logging.error(f"Could not read input directory {input_dir}: {e}")
        return

    # Iterate through items found in the input directory
    for item in items_in_dir:
        if item.is_file() and item.suffix.lower() == ".txt":
            if item in processed_files:
                continue # Skip if already processed in this run

            output_file_path = output_dir / item.name # Define corresponding output path
            # Convert the individual file
            convert_file(item, output_file_path)
            processed_files.add(item) # Mark as done
            converted_count += 1
        elif not item.is_file(): # It's a directory
            skipped_count +=1
        else: # It's a file but not .txt
            skipped_count +=1

    # Log summary for this directory conversion
    logging.info(f"--- Conversion Complete for this directory ---")
    logging.info(f"Processed and attempted conversion for {converted_count} .txt files.")
    if skipped_count > 0:
         logging.info(f"Skipped {skipped_count} items (non-.txt files or directories).")
    logging.info(f"Converted labels saved to: {output_dir}")

# --- Main Execution Block (runs if script is executed directly) ---
if __name__ == "__main__":
    # Setup command-line argument parsing
    parser = argparse.ArgumentParser(description="Convert YOLO segmentation labels to detection labels.")
    parser.add_argument("input_dir", type=str, help="Directory containing input segmentation label files (.txt)")
    parser.add_argument("output_dir", type=str, help="Directory where output detection label files will be saved")
    args = parser.parse_args() # Parse arguments from command line

    # Convert string paths from arguments to Path objects
    INPUT_LABEL_DIR = Path(args.input_dir)
    OUTPUT_LABEL_DIR = Path(args.output_dir)

    # Start the conversion process for the specified directories
    logging.info(f"Starting conversion from '{INPUT_LABEL_DIR}' to '{OUTPUT_LABEL_DIR}'")
    convert_directory(INPUT_LABEL_DIR, OUTPUT_LABEL_DIR)
"""

# --- Write the script content to a file in the Colab environment ---
try:
    # Open the file in write mode ('w') which creates it if it doesn't exist
    with open("convert_labels.py", "w") as f:
        # Write the script content, removing any leading/trailing whitespace from the multi-line string
        f.write(script_content.strip())
    print("Successfully saved 'convert_labels.py'")
except Exception as e:
    # Print an error message if saving the file fails
    print(f"Error saving script: {e}")


```

```python
import os
import shutil
import glob
import logging
from pathlib import Path

# Setup logging
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')

# --- Configuration ---

# 1. Path to the EXISTING dataset FOLDER TO BE MODIFIED (Originally 19 classes)
target_dataset_base_dir = Path('/content/drive/MyDrive/Combine_dataset') # <--- VERIFY THIS

# 2. Path to the NEW dataset BASE FOLDER (Lane dataset) to be added
#    The script will look for 'train_seg', 'valid_seg', 'test_seg' inside this path.
new_dataset_base_dir = Path('/content/drive/MyDrive/Lane') # <--- VERIFY THIS

# 3. Class ID Mapping: How to renumber classes from the NEW dataset
#    Format: { original_id_in_new_dataset : desired_id_in_combined_dataset }
#    Your Lane dataset had 0='dotted-line', 1='line'.
#    Your Combine dataset has 0-18. So, new IDs should be 19 and 20.
class_id_mapping = {
    0: 19, # Original ID 0 (dotted-line) will become ID 19
    1: 20  # Original ID 1 (line) will become ID 20
}
# Calculate the expected final number of classes and max ID
original_target_nc = 19 # Number of classes originally in the target dataset
new_classes_added = len(class_id_mapping)
final_nc = original_target_nc + new_classes_added
final_max_id = final_nc - 1

# --- Dataset Structure Settings ---
# Does your TARGET (Combine_dataset) dataset use 'images' and 'labels' subfolders within train/valid/test?
use_images_labels_subfolders_TARGET = True # Set to True or False

# --- Filename Conflict Handling ---
# If True, rename files from the new dataset (prefix 'lane_') if a file
# with the same name already exists in the target dataset. Prevents overwriting.
# If False, files with the same name WILL BE OVERWRITTEN (Use with caution!).
rename_new_files_on_conflict = True # Set to False ONLY if you understand the risk
rename_prefix = "lane_" # Prefix to use when renaming conflicting files

# --- End Configuration ---

# --- Print configuration for verification ---
print("--- Configuration Summary (In-Place Merge) ---")
print(f"Target (Existing) Dataset Path: {target_dataset_base_dir} <--- WILL BE MODIFIED")
print(f"  Structure uses images/labels subfolders: {use_images_labels_subfolders_TARGET}")
print(f"New Dataset Path (Base): {new_dataset_base_dir}")
# *** FIXED PRINT STATEMENT BELOW ***
print(f"  Script expects '<split_name>_seg' subfolders (e.g., train_seg) inside the New Dataset Path.")
print(f"  Script expects 'images' and 'labels_detect' subfolders inside each '<split_name>_seg' folder.")
# *** END OF FIX ***
print(f"Class ID Mapping (Original New ID -> Final Combined ID): {class_id_mapping}")
print(f"Expected Final Class Count (nc): {final_nc}")
print(f"Expected Final Max Class ID: {final_max_id}")
print(f"Rename new files on conflict: {rename_new_files_on_conflict}")
if rename_new_files_on_conflict:
    print(f"  Rename prefix: '{rename_prefix}'")
print("-----------------------------")
print(" WARNING: The TARGET dataset folder listed above will be directly modified. ")
print(" It is strongly recommended to BACK UP the target folder before proceeding. ")
print("-----------------------------")

# --- Basic Path Validation ---
if not target_dataset_base_dir.is_dir():
    logging.critical(f"!!! CRITICAL ERROR: Target dataset directory NOT FOUND: {target_dataset_base_dir}")
    # sys.exit("Target dataset not found.") # Uncomment to force exit
if not new_dataset_base_dir.is_dir():
    logging.warning(f"!!! WARNING: New dataset directory NOT FOUND: {new_dataset_base_dir}")
    # sys.exit("New dataset not found.") # Uncomment to force exit

# --- Function to process a single label file ---
def process_label_file(lbl_source_path, lbl_dest_path, mapping):
    """Reads a label file, modifies class IDs based on mapping, writes to dest."""
    modified_lines = []
    found_valid_line = False
    try:
        with open(lbl_source_path, 'r') as infile:
            for line_num, line in enumerate(infile, 1):
                parts = line.strip().split()
                if len(parts) >= 5: # Basic check for YOLO format (class + 4 coords)
                    try:
                        original_class_id = int(parts[0])
                        if original_class_id in mapping:
                            parts[0] = str(mapping[original_class_id]) # Replace with new ID
                            modified_lines.append(" ".join(parts))
                            found_valid_line = True
                        else:
                            # This case should ideally not happen if the new dataset is clean
                            logging.warning(f"Found unexpected class ID {original_class_id} in new dataset file '{lbl_source_path.name}', line {line_num+1}. Keeping original ID (check mapping).")
                            modified_lines.append(line.strip()) # Keep original line
                            found_valid_line = True
                    except ValueError:
                        logging.warning(f"Non-integer class ID in '{lbl_source_path.name}', line {line_num+1}. Skipping line: '{line.strip()}'")
                elif line.strip(): # Line is not empty but doesn't have enough parts
                    logging.warning(f"Malformed line in '{lbl_source_path.name}', line {line_num+1}. Skipping line: '{line.strip()}'")

        # Write modified label file to TARGET destination only if valid lines were processed
        if found_valid_line:
            lbl_dest_path.parent.mkdir(parents=True, exist_ok=True) # Ensure dir exists
            with open(lbl_dest_path, 'w') as outfile:
                outfile.write("\n".join(modified_lines) + "\n")
            return True # Indicate success
        else:
            logging.warning(f"No valid lines found or modified in '{lbl_source_path.name}'. Output file '{lbl_dest_path.name}' not created.")
            return False # Indicate no valid lines processed

    except Exception as e:
        logging.error(f"ERROR processing label file '{lbl_source_path.name}': {e}")
        return False # Indicate error

# --- Main Processing Loop ---
print("\nPlease carefully review the paths, settings, and warnings above.")
print("--- Starting: Processing and ADDING NEW Dataset Files to Target Dataset ---")

if target_dataset_base_dir.is_dir() and new_dataset_base_dir.is_dir(): # Proceed only if both base dirs exist
    for split in ['train', 'valid', 'test']:
        print(f"\nProcessing NEW dataset split: {split} -> adding to TARGET '{split}'")

        # *** MODIFIED PATH CONSTRUCTION FOR LANE DATASET ***
        # Construct the specific directory name for the new dataset (e.g., train_seg)
        new_split_dir_name = f"{split}_seg"
        new_split_dir = new_dataset_base_dir / new_split_dir_name # e.g., /content/drive/MyDrive/Lane/train_seg

        # Determine NEW dataset source paths based on the _seg structure
        # Images are likely still in an 'images' subfolder within the _seg folder
        new_img_src_dir = new_split_dir / 'images' # e.g., /content/drive/MyDrive/Lane/train_seg/images
        # Labels are now in the 'labels_detect' subfolder within the _seg folder (from conversion)
        new_lbl_src_dir = new_split_dir / 'labels_detect' # e.g., /content/drive/MyDrive/Lane/train_seg/labels_detect
        # *** END OF MODIFIED PATH CONSTRUCTION ***


        # Determine TARGET dataset destination paths (where files will be added/modified)
        target_img_dest_dir = target_dataset_base_dir / split / 'images' if use_images_labels_subfolders_TARGET else target_dataset_base_dir / split
        target_lbl_dest_dir = target_dataset_base_dir / split / 'labels' if use_images_labels_subfolders_TARGET else target_dataset_base_dir / split

        processed_img_count = 0
        processed_lbl_count = 0
        skipped_conflict_count = 0
        overwritten_count = 0
        error_count = 0

        # Ensure target destination directories exist (create if necessary)
        try:
            target_img_dest_dir.mkdir(parents=True, exist_ok=True)
            target_lbl_dest_dir.mkdir(parents=True, exist_ok=True)
        except OSError as e:
             logging.error(f"Could not create target directories for split {split}: {e}. Skipping split.")
             continue

        # Check if source directories exist for the new dataset's split
        if not new_lbl_src_dir.is_dir():
            logging.info(f"  Info: NEW label source directory not found for split '{split}', nothing to add from here: {new_lbl_src_dir}")
            continue # Skip this split if no source labels folder
        if not new_img_src_dir.is_dir():
            logging.warning(f"  Warning: NEW image source directory not found for split '{split}', will process labels but not copy images: {new_img_src_dir}")
            # This allows processing labels even if images are missing, but it's unusual.

        print(f"  Processing labels from '{new_lbl_src_dir}'...")
        # Iterate through label files first
        try:
            label_files_in_new = list(new_lbl_src_dir.glob('*.txt'))
            logging.info(f"  Found {len(label_files_in_new)} .txt files in source label directory.")
        except Exception as e:
            logging.error(f"  Could not list files in source label directory {new_lbl_src_dir}: {e}")
            continue # Skip split if cannot read source labels

        for lbl_source_path in label_files_in_new:
            base_name = lbl_source_path.stem # Filename without extension

            # Find corresponding image file in the source
            img_filename = None
            img_source_path = None
            if new_img_src_dir.is_dir(): # Only look for images if src dir exists
                for ext in ['.png', '.jpg', '.jpeg', '.bmp', '.webp']: # Added more extensions
                    potential_img_path = new_img_src_dir / (base_name + ext)
                    if potential_img_path.exists():
                        img_filename = potential_img_path.name
                        img_source_path = potential_img_path
                        break

            if not img_source_path and new_img_src_dir.is_dir():
                logging.warning(f"    Warning: No corresponding image found for label '{lbl_source_path.name}' in source. Skipping label processing for this file.")
                error_count += 1
                continue
            elif not new_img_src_dir.is_dir():
                logging.info(f"    Info: Processing label '{lbl_source_path.name}' without copying image (source image dir missing).")
                img_filename = None # Ensure we don't try to copy image later

            # --- Determine Destination Filenames & Handle Conflicts ---
            dest_lbl_filename = lbl_source_path.name
            dest_img_filename = img_filename # Can be None

            lbl_dest_path = target_lbl_dest_dir / dest_lbl_filename
            img_dest_path = target_img_dest_dir / dest_img_filename if dest_img_filename else None

            # Check for conflicts
            conflict_exists = lbl_dest_path.exists() or (img_dest_path and img_dest_path.exists())

            if conflict_exists:
                if rename_new_files_on_conflict:
                    logging.info(f"    Info: Conflict for '{lbl_source_path.name}'/'{img_filename}'. Renaming with '{rename_prefix}' prefix.")
                    original_lbl_filename = dest_lbl_filename
                    original_img_filename = dest_img_filename

                    # Create new prefixed names
                    dest_lbl_filename = f"{rename_prefix}{original_lbl_filename}"
                    dest_img_filename = f"{rename_prefix}{original_img_filename}" if original_img_filename else None

                    # Update destination paths
                    lbl_dest_path = target_lbl_dest_dir / dest_lbl_filename
                    img_dest_path = target_img_dest_dir / dest_img_filename if dest_img_filename else None

                    # Check again if even the renamed file exists
                    if lbl_dest_path.exists() or (img_dest_path and img_dest_path.exists()):
                        logging.warning(f"    Warning: Renamed file '{dest_lbl_filename}'/'{dest_img_filename}' also exists. Skipping.")
                        skipped_conflict_count += 1
                        continue
                else: # rename_new_files_on_conflict is False, so we will overwrite
                    logging.warning(f"    Warning: Overwriting existing file(s) for '{lbl_source_path.name}'/'{img_filename}' (rename_on_conflict=False).")
                    overwritten_count += 1 # Technically overwriting label and maybe image

            # --- Process Label File (Change Class IDs using mapping) ---
            label_processed_ok = process_label_file(lbl_source_path, lbl_dest_path, class_id_mapping)

            if label_processed_ok:
                processed_lbl_count += 1
                # Copy corresponding image file to TARGET destination (if it exists and path is set)
                if img_source_path and img_dest_path:
                    try:
                        shutil.copy2(img_source_path, img_dest_path) # copy2 preserves metadata
                        processed_img_count += 1
                    except Exception as e:
                         logging.error(f"    ERROR copying image '{img_source_path.name}' to '{img_dest_path}': {e}")
                         error_count += 1
                         # Consider deleting the just-written label file if image copy fails?
                         # if lbl_dest_path.exists(): lbl_dest_path.unlink()
                         # processed_lbl_count -= 1
                elif not img_dest_path and img_filename:
                    # Image source dir was missing, label processed, image skipped as noted earlier
                    pass
            else:
                # Label processing failed (error logged in function)
                error_count += 1


        print(f"  Split '{split}' summary:")
        print(f"    Labels Processed/Written to Target: {processed_lbl_count}")
        print(f"    Images Copied to Target: {processed_img_count}")
        print(f"    Skipped (due to conflicts/existing renamed): {skipped_conflict_count}")
        print(f"    Overwritten (rename_on_conflict=False): {overwritten_count}")
        print(f"    Errors/Warnings during processing: {error_count}")

    print("\n--- Finished: Processing and ADDING NEW Dataset Files to Target Dataset ---")

else:
    print("--- Skipping Processing: One or both dataset base directories not found. ---")


# --- Verification Step 1: File Counts ---
# (Verification code remains the same as previous version)
print("\n--- Verification: File Counts in MODIFIED Target Dataset ---")
print(f"Checking folder: {target_dataset_base_dir}")

total_images = 0
total_labels = 0

if target_dataset_base_dir.is_dir(): # Only verify if target exists
    for split in ['train', 'valid', 'test']:
        print(f"\nVerifying split: {split}")
        split_dir = target_dataset_base_dir / split

        # Determine target paths based on target structure setting
        img_folder = split_dir / 'images' if use_images_labels_subfolders_TARGET else split_dir
        lbl_folder = split_dir / 'labels' if use_images_labels_subfolders_TARGET else split_dir

        # Count image files
        img_count = 0
        if img_folder.is_dir():
            img_files = list(img_folder.glob('*.jpg')) + \
                        list(img_folder.glob('*.jpeg')) + \
                        list(img_folder.glob('*.png')) + \
                        list(img_folder.glob('*.bmp')) + \
                        list(img_folder.glob('*.webp'))
            img_count = len(img_files)
            print(f"  Image files found in '{img_folder}': {img_count}")
            total_images += img_count
        else:
            print(f"  Image folder not found: {img_folder}")

        # Count label files
        lbl_count = 0
        if lbl_folder.is_dir():
            lbl_files = list(lbl_folder.glob('*.txt'))
            lbl_count = len(lbl_files)
            print(f"  Label files (.txt) found in '{lbl_folder}': {lbl_count}")
            total_labels += lbl_count
        else:
            print(f"  Label folder not found: {lbl_folder}")

        if img_count != lbl_count:
            print(f"  Warning: Mismatch between image count ({img_count}) and label count ({lbl_count}) in split '{split}'.")

    print("\n-----------------------------")
    print(f"Total Target Images Now: {total_images}")
    print(f"Total Target Labels Now: {total_labels}")
    if total_images != total_labels:
        print("Warning: Overall image and label counts do not match!")
    print("-----------------------------")
    print("Note: Counts should now be higher than the original dataset.")
else:
    print("Target directory not found, skipping file count verification.")


# --- Verification Step 2: Class IDs ---
# (Verification code remains the same as previous version)
print("\n--- Verification: Class IDs in MODIFIED Target Dataset Labels ---")

expected_min_id = 0
expected_max_id = final_max_id # Calculated based on config
expected_total_classes_min = final_nc # Expect at least this many unique IDs overall

all_splits_ok = True
overall_unique_ids = set()

if target_dataset_base_dir.is_dir(): # Only verify if target exists
    for split in ['train', 'valid', 'test']:
        print(f"\nVerifying split: {split}")
        split_dir = target_dataset_base_dir / split

        # Determine target label path
        lbl_folder = split_dir / 'labels' if use_images_labels_subfolders_TARGET else split_dir

        if not lbl_folder.is_dir():
            print(f"  Label folder not found, skipping check: {lbl_folder}")
            # If this split *should* have labels, it's an issue.
            # You might want to set all_splits_ok = False here depending on expectations
            continue

        unique_class_ids_split = set()
        processed_file_count = 0
        total_txt_files = 0
        try:
            files_in_dir = list(lbl_folder.glob('*.txt'))
            total_txt_files = len(files_in_dir)
        except Exception as e:
            logging.error(f"  Could not list files in target label directory {lbl_folder}: {e}")
            continue # Skip split verification if cannot read target labels

        for file_path in files_in_dir:
            try:
                with open(file_path, 'r') as f:
                    for line_num, line in enumerate(f):
                        parts = line.strip().split()
                        if len(parts) > 0:
                            try:
                                class_id = int(parts[0])
                                unique_class_ids_split.add(class_id)
                                overall_unique_ids.add(class_id) # Add to overall set
                            except ValueError:
                                logging.warning(f"    Warning: Non-integer ID in '{file_path.name}', line {line_num + 1}: '{line.strip()}'")
                processed_file_count += 1
            except Exception as e:
                logging.error(f"    Error reading file '{file_path.name}': {e}")

        print(f"  Scan complete. Processed {processed_file_count} out of {total_txt_files} '.txt' files found.")

        if not unique_class_ids_split:
            print("  No valid class IDs found in this split.")
            if total_txt_files > 0 : all_splits_ok = False # If txt files exist but no IDs found, it's likely an issue
        else:
            sorted_ids = sorted(list(unique_class_ids_split))
            min_id = min(sorted_ids)
            max_id = max(sorted_ids)
            num_classes = len(sorted_ids)

            print(f"  Unique Class IDs found in this split: {sorted_ids}")
            print(f"  Min ID: {min_id}, Max ID: {max_id}, Count: {num_classes}")

            # Check against expectations for this split
            split_ok = True
            if min_id != expected_min_id:
                print(f"    Warning: Minimum ID in this split is {min_id}, expected {expected_min_id}.")
                split_ok = False
            if max_id > expected_max_id:
                 print(f"    Warning: Maximum ID in this split is {max_id}, expected no more than {expected_max_id}.")
                 split_ok = False
            # Check if the *new* class IDs are present (they might not be in every split, especially test/valid)
            new_ids_in_split = {id for id in class_id_mapping.values() if id in unique_class_ids_split}
            if not new_ids_in_split:
                 print(f"    Info: None of the newly mapped class IDs {list(class_id_mapping.values())} were found in this split's labels.")
            else:
                 print(f"    Info: Found newly mapped class IDs in this split: {sorted(list(new_ids_in_split))}")


            if not split_ok:
                all_splits_ok = False # Mark overall as not OK if specific checks fail

    # --- Overall Class ID Check ---
    print("\n--- Overall Class ID Check Summary (Across All Splits) ---")
    if not overall_unique_ids:
         print("No valid class IDs found across any splits.")
         all_splits_ok = False
    else:
        sorted_overall_ids = sorted(list(overall_unique_ids))
        min_overall_id = min(sorted_overall_ids)
        max_overall_id = max(sorted_overall_ids)
        count_overall_ids = len(sorted_overall_ids)
        print(f"Overall Unique IDs found: {sorted_overall_ids}")
        print(f"Min: {min_overall_id}, Max: {max_overall_id}, Count: {count_overall_ids}")

        if min_overall_id != expected_min_id:
            print(f"Warning: Overall Minimum ID is {min_overall_id}, expected {expected_min_id}.")
            all_splits_ok = False
        if max_overall_id != expected_max_id:
            print(f"Warning: Overall Maximum ID is {max_overall_id}, expected {expected_max_id}.")
            all_splits_ok = False
        if count_overall_ids != expected_total_classes_min:
             print(f"Warning: Found {count_overall_ids} unique classes overall, expected {expected_total_classes_min}.")
             all_splits_ok = False

        # Check if all new IDs are present *somewhere* in the dataset
        all_new_ids_present = all(new_id in overall_unique_ids for new_id in class_id_mapping.values())
        if not all_new_ids_present:
             print(f"Warning: Not all newly mapped class IDs {list(class_id_mapping.values())} were found across the entire dataset!")
             all_splits_ok = False

        if all_splits_ok:
             print("Overall class ID range and count appear correct.")
        else:
             print("Potential issues found with overall class ID range or count. Please review warnings.")

else:
    print("Target directory not found, skipping class ID verification.")

print("------------------------------------")

```

```python
# -*- coding: utf-8 -*-
# This cell writes the conversion script to a file named 'convert_labels.py'

import os
from pathlib import Path
import logging
import argparse # For command-line arguments

# --- Script Content ---
# Defines the Python code that will be written to the file
script_content = """
# -*- coding: utf-8 -*-
import os
from pathlib import Path
import logging
import argparse # For command-line arguments

# Setup logging
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')

def convert_file(input_path: Path, output_path: Path):
    \"\"\"
    Converts a single segmentation label file to a detection label file.

    Args:
        input_path: Path to the input segmentation label file (.txt).
        output_path: Path where the output detection label file (.txt) will be saved.
    \"\"\"
    try:
        output_lines = []
        # Open and read the input segmentation file line by line
        with open(input_path, 'r') as infile:
            for line_num, line in enumerate(infile, 1):
                parts = line.strip().split() # Split line into parts
                if not parts:
                    continue # Skip empty lines

                # Check for minimum required parts (class_id + one point) and odd number of coordinate values
                if len(parts) < 3 or len(parts) % 2 == 0:
                    logging.warning(f"Skipping malformed line {line_num} in {input_path}: Incorrect number of parts ({len(parts)}). Line: '{line.strip()}'")
                    continue

                try:
                    # Extract class ID and convert coordinates to floats
                    class_id = int(parts[0])
                    coords = [float(p) for p in parts[1:]]
                except ValueError:
                    # Handle cases where coordinate values are not numbers
                    logging.warning(f"Skipping malformed line {line_num} in {input_path}: Non-numeric value found. Line: '{line.strip()}'")
                    continue

                # Separate x and y coordinates from the list
                x_coords = coords[0::2] # Get elements at even indices (0, 2, 4, ...)
                y_coords = coords[1::2] # Get elements at odd indices (1, 3, 5, ...)

                # Ensure coordinates were actually extracted
                if not x_coords or not y_coords:
                     logging.warning(f"Skipping line {line_num} in {input_path}: No valid coordinates found after parsing. Line: '{line.strip()}'")
                     continue

                # Find the min/max x and y values to define the bounding box
                min_x = min(x_coords)
                max_x = max(x_coords)
                min_y = min(y_coords)
                max_y = max(y_coords)

                # Calculate YOLO detection format parameters (normalized)
                center_x = (min_x + max_x) / 2.0
                center_y = (min_y + max_y) / 2.0
                width = max_x - min_x
                height = max_y - min_y

                # Validate calculated bounding box values
                if not (0 <= center_x <= 1 and 0 <= center_y <= 1 and width >= 0 and height >= 0):
                     logging.warning(f"Skipping line {line_num} in {input_path}: Calculated bbox invalid or outside bounds (center=[{center_x:.4f},{center_y:.4f}], w={width:.4f}, h={height:.4f}). Original line: '{line.strip()}'")
                     continue
                # Log if box has zero width or height (might be intended or an issue)
                if width == 0 or height == 0:
                    logging.warning(f"Line {line_num} in {input_path} resulted in zero width or height (w={width:.4f}, h={height:.4f}). Keeping it, but check if this is intended. Original line: '{line.strip()}'")

                # Format the output line string
                output_line = f"{class_id} {center_x:.6f} {center_y:.6f} {width:.6f} {height:.6f}"
                output_lines.append(output_line) # Add to list of lines for this file

        # Write the converted detection labels to the output file
        output_path.parent.mkdir(parents=True, exist_ok=True) # Ensure output directory exists
        if output_lines: # Only create/write file if there are valid lines
            with open(output_path, 'w') as outfile:
                outfile.write("\\n".join(output_lines) + "\\n") # Join lines with newline and add trailing newline
        elif not output_path.exists():
             # Create an empty file if input was empty/invalid, prevents missing label errors later
             output_path.touch()

    except FileNotFoundError:
        logging.error(f"Input file not found: {input_path}")
    except Exception as e:
        logging.error(f"An unexpected error occurred processing {input_path}: {e}")

def convert_directory(input_dir: Path, output_dir: Path):
    \"\"\"
    Converts all .txt files in an input directory from segmentation to detection format.

    Args:
        input_dir: Directory containing the input segmentation label files (.txt).
        output_dir: Directory where the output detection label files (.txt) will be saved.
    \"\"\"
    if not input_dir.is_dir():
        logging.error(f"Input directory not found or is not a directory: {input_dir}")
        return

    output_dir.mkdir(parents=True, exist_ok=True) # Create output dir if needed

    converted_count = 0
    skipped_count = 0
    processed_files = set() # To avoid reprocessing if run multiple times

    logging.info(f"Scanning input directory: {input_dir}")
    try:
        items_in_dir = list(input_dir.iterdir()) # Get directory contents
        logging.info(f"Found {len(items_in_dir)} items in input directory.")
    except Exception as e:
        logging.error(f"Could not read input directory {input_dir}: {e}")
        return

    # Iterate through items found in the input directory
    for item in items_in_dir:
        if item.is_file() and item.suffix.lower() == ".txt":
            if item in processed_files:
                continue # Skip if already processed in this run

            output_file_path = output_dir / item.name # Define corresponding output path
            # Convert the individual file
            convert_file(item, output_file_path)
            processed_files.add(item) # Mark as done
            converted_count += 1
        elif not item.is_file(): # It's a directory
            skipped_count +=1
        else: # It's a file but not .txt
            skipped_count +=1

    # Log summary for this directory conversion
    logging.info(f"--- Conversion Complete for this directory ---")
    logging.info(f"Processed and attempted conversion for {converted_count} .txt files.")
    if skipped_count > 0:
         logging.info(f"Skipped {skipped_count} items (non-.txt files or directories).")
    logging.info(f"Converted labels saved to: {output_dir}")

# --- Main Execution Block (runs if script is executed directly) ---
if __name__ == "__main__":
    # Setup command-line argument parsing
    parser = argparse.ArgumentParser(description="Convert YOLO segmentation labels to detection labels.")
    parser.add_argument("input_dir", type=str, help="Directory containing input segmentation label files (.txt)")
    parser.add_argument("output_dir", type=str, help="Directory where output detection label files will be saved")
    args = parser.parse_args() # Parse arguments from command line

    # Convert string paths from arguments to Path objects
    INPUT_LABEL_DIR = Path(args.input_dir)
    OUTPUT_LABEL_DIR = Path(args.output_dir)

    # Start the conversion process for the specified directories
    logging.info(f"Starting conversion from '{INPUT_LABEL_DIR}' to '{OUTPUT_LABEL_DIR}'")
    convert_directory(INPUT_LABEL_DIR, OUTPUT_LABEL_DIR)
"""

# --- Write the script content to a file in the Colab environment ---
try:
    # Open the file in write mode ('w') which creates it if it doesn't exist
    with open("convert_labels.py", "w") as f:
        # Write the script content, removing any leading/trailing whitespace from the multi-line string
        f.write(script_content.strip())
    print("Successfully saved 'convert_labels.py'")
except Exception as e:
    # Print an error message if saving the file fails
    print(f"Error saving script: {e}")

```

```python
import os
import shutil
import glob
import logging
from pathlib import Path
import sys # Import sys for exit

# Setup logging
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')

# --- Configuration ---

# 1. Path to the EXISTING dataset FOLDER TO BE MODIFIED (Should have 21 classes, 0-20)
target_dataset_base_dir = Path('/content/drive/MyDrive/Combine_dataset') # <--- VERIFY THIS

# 2. Path to the NEW dataset containing 'blackline' examples (with converted labels)
#    This folder should contain train/, valid/, test/ subfolders
new_dataset_base_dir = Path('/content/drive/MyDrive/BlackLine2') # <--- EDIT THIS PATH if different

# 3. Class ID Mapping: Map the 'blackline' class ID from the NEW dataset to the EXISTING ID 20
#    Format: { original_id_in_blackline_dataset : desired_id_in_combined_dataset }
#    *** EDIT the key (0) if 'blackline' had a different ID in its original labels ***
class_id_mapping = {
    0: 20  # Assuming 'blackline' was class ID 0 originally, map it to ID 20
}
# Expected final number of classes and max ID DO NOT CHANGE
final_nc = 21
final_max_id = 20

# --- Dataset Structure Settings ---
# Does your TARGET (Combine_dataset) dataset use 'images' and 'labels' subfolders within train/valid/test?
use_images_labels_subfolders_TARGET = True # Set based on Combine_dataset structure

# --- Filename Conflict Handling ---
rename_new_files_on_conflict = True # Keep True (safer)
rename_prefix = "blackline_" # Changed prefix to distinguish these files

# --- End Configuration ---

# --- Print configuration for verification ---
print("--- Configuration Summary (Adding More Blackline Data to Class 20) ---")
print(f"Target (Existing) Dataset Path: {target_dataset_base_dir} <--- WILL BE MODIFIED")
print(f"  Structure uses images/labels subfolders: {use_images_labels_subfolders_TARGET}")
print(f"New Dataset Path (Blackline): {new_dataset_base_dir}")
print(f"  Script expects 'images' and 'labels_detect' subfolders inside each train/valid/test folder of the New Dataset.")
print(f"Class ID Mapping (Original Blackline ID -> Final Combined ID): {class_id_mapping}")
print(f"Expected Final Class Count (nc): {final_nc}") # Should remain 21
print(f"Expected Final Max Class ID: {final_max_id}") # Should remain 20
print(f"Rename new files on conflict: {rename_new_files_on_conflict}")
if rename_new_files_on_conflict:
    print(f"  Rename prefix: '{rename_prefix}'")
print("-----------------------------")
print(" WARNING: The TARGET dataset folder listed above will be directly modified. ")
print(" It is strongly recommended to BACK UP the target folder before proceeding. ")
print("-----------------------------")

# --- Basic Path Validation ---
if not target_dataset_base_dir.is_dir():
    logging.critical(f"!!! CRITICAL ERROR: Target dataset directory NOT FOUND: {target_dataset_base_dir}")
    sys.exit("Target dataset not found.") # Force exit
if not new_dataset_base_dir.is_dir():
    logging.critical(f"!!! CRITICAL ERROR: New dataset directory NOT FOUND: {new_dataset_base_dir}")
    sys.exit("New dataset not found.") # Force exit

# --- Function to process a single label file (same as before) ---
def process_label_file(lbl_source_path, lbl_dest_path, mapping):
    """Reads a label file, modifies class IDs based on mapping, writes to dest."""
    modified_lines = []
    found_valid_line = False
    try:
        with open(lbl_source_path, 'r') as infile:
            for line_num, line in enumerate(infile, 1):
                parts = line.strip().split()
                if len(parts) >= 5: # Basic check for YOLO format (class + 4 coords)
                    try:
                        original_class_id = int(parts[0])
                        if original_class_id in mapping:
                            # Map the class ID to the target ID (e.g., 20)
                            parts[0] = str(mapping[original_class_id])
                            modified_lines.append(" ".join(parts))
                            found_valid_line = True
                        else:
                            # Log if the new dataset contains IDs other than the one we want to map
                            logging.warning(f"Found unexpected class ID {original_class_id} in new dataset file '{lbl_source_path.name}', line {line_num+1}. Skipping this line as it's not in the mapping: {mapping}")
                            # Do NOT append this line to modified_lines unless you want to keep other classes too
                    except ValueError:
                        logging.warning(f"Non-integer class ID in '{lbl_source_path.name}', line {line_num+1}. Skipping line: '{line.strip()}'")
                elif line.strip(): # Line is not empty but doesn't have enough parts
                    logging.warning(f"Malformed line in '{lbl_source_path.name}', line {line_num+1}. Skipping line: '{line.strip()}'")

        # Write modified label file to TARGET destination only if valid lines were processed
        if found_valid_line:
            lbl_dest_path.parent.mkdir(parents=True, exist_ok=True) # Ensure dir exists
            with open(lbl_dest_path, 'w') as outfile:
                outfile.write("\n".join(modified_lines) + "\n")
            return True # Indicate success
        else:
            # Log if the file contained only classes we didn't map or was empty/malformed
            logging.info(f"No lines mapped or file invalid in '{lbl_source_path.name}'. Output file '{lbl_dest_path.name}' not created/updated.")
            return False # Indicate no lines relevant to the mapping were processed

    except Exception as e:
        logging.error(f"ERROR processing label file '{lbl_source_path.name}': {e}")
        return False # Indicate error


# --- Main Processing Loop ---
print("\nPlease carefully review the paths, settings, and warnings above.")
print("--- Starting: Processing and ADDING NEW Blackline Dataset Files to Target Dataset ---")

if target_dataset_base_dir.is_dir() and new_dataset_base_dir.is_dir(): # Proceed only if both base dirs exist
    for split in ['train', 'valid', 'test']:
        print(f"\nProcessing NEW dataset split: {split} -> adding to TARGET '{split}'")

        # Determine NEW dataset source paths (Blackline with converted labels)
        new_split_dir = new_dataset_base_dir / split
        # Source Images: <new_dataset_base_dir>/<split>/images
        new_img_src_dir = new_split_dir / 'images'
        # Source Labels: <new_dataset_base_dir>/<split>/labels_detect (using converted labels)
        new_lbl_src_dir = new_split_dir / 'labels' # <<<--- Using converted labels path

        # Determine TARGET dataset destination paths (where files will be added/modified)
        # Target Images: <target_dataset_base_dir>/<split>/images
        target_img_dest_dir = target_dataset_base_dir / split / 'images' if use_images_labels_subfolders_TARGET else target_dataset_base_dir / split
        # Target Labels: <target_dataset_base_dir>/<split>/labels
        target_lbl_dest_dir = target_dataset_base_dir / split / 'labels' if use_images_labels_subfolders_TARGET else target_dataset_base_dir / split

        processed_img_count = 0
        processed_lbl_count = 0
        skipped_conflict_count = 0
        overwritten_count = 0
        error_count = 0

        # Ensure target destination directories exist (create if necessary)
        try:
            target_img_dest_dir.mkdir(parents=True, exist_ok=True)
            target_lbl_dest_dir.mkdir(parents=True, exist_ok=True)
        except OSError as e:
             logging.error(f"Could not create target directories for split {split}: {e}. Skipping split.")
             continue

        # Check if source directories exist for the new dataset's split
        if not new_lbl_src_dir.is_dir():
            logging.info(f"  Info: NEW label source directory not found for split '{split}', nothing to add from here: {new_lbl_src_dir}")
            continue # Skip this split if no source labels folder
        if not new_img_src_dir.is_dir():
            logging.warning(f"  Warning: NEW image source directory not found for split '{split}', will process labels but not copy images: {new_img_src_dir}")

        print(f"  Processing labels from '{new_lbl_src_dir}'...")
        # Iterate through label files first
        try:
            label_files_in_new = list(new_lbl_src_dir.glob('*.txt'))
            logging.info(f"  Found {len(label_files_in_new)} .txt files in source label directory.")
        except Exception as e:
            logging.error(f"  Could not list files in source label directory {new_lbl_src_dir}: {e}")
            continue # Skip split if cannot read source labels

        for lbl_source_path in label_files_in_new:
            base_name = lbl_source_path.stem # Filename without extension

            # Find corresponding image file in the source image directory
            img_filename = None
            img_source_path = None
            if new_img_src_dir.is_dir(): # Only look for images if src dir exists
                for ext in ['.png', '.jpg', '.jpeg', '.bmp', '.webp']:
                    potential_img_path = new_img_src_dir / (base_name + ext)
                    if potential_img_path.exists():
                        img_filename = potential_img_path.name
                        img_source_path = potential_img_path
                        break

            if not img_source_path and new_img_src_dir.is_dir():
                logging.warning(f"    Warning: No corresponding image found for label '{lbl_source_path.name}' in source image dir '{new_img_src_dir}'. Skipping label processing for this file.")
                error_count += 1
                continue
            elif not new_img_src_dir.is_dir():
                logging.info(f"    Info: Processing label '{lbl_source_path.name}' without copying image (source image dir missing).")
                img_filename = None # Ensure we don't try to copy image later

            # --- Determine Destination Filenames & Handle Conflicts ---
            dest_lbl_filename = lbl_source_path.name
            dest_img_filename = img_filename # Can be None

            lbl_dest_path = target_lbl_dest_dir / dest_lbl_filename
            img_dest_path = target_img_dest_dir / dest_img_filename if dest_img_filename else None

            # Check for conflicts
            conflict_exists = lbl_dest_path.exists() or (img_dest_path and img_dest_path.exists())

            if conflict_exists:
                if rename_new_files_on_conflict:
                    logging.info(f"    Info: Conflict for '{lbl_source_path.name}'/'{img_filename}'. Renaming with '{rename_prefix}' prefix.")
                    original_lbl_filename = dest_lbl_filename
                    original_img_filename = dest_img_filename

                    # Create new prefixed names
                    dest_lbl_filename = f"{rename_prefix}{original_lbl_filename}"
                    dest_img_filename = f"{rename_prefix}{original_img_filename}" if original_img_filename else None

                    # Update destination paths
                    lbl_dest_path = target_lbl_dest_dir / dest_lbl_filename
                    img_dest_path = target_img_dest_dir / dest_img_filename if dest_img_filename else None

                    # Check again if even the renamed file exists
                    if lbl_dest_path.exists() or (img_dest_path and img_dest_path.exists()):
                        logging.warning(f"    Warning: Renamed file '{dest_lbl_filename}'/'{dest_img_filename}' also exists. Skipping.")
                        skipped_conflict_count += 1
                        continue
                else: # rename_new_files_on_conflict is False, so we will overwrite
                    logging.warning(f"    Warning: Overwriting existing file(s) for '{lbl_source_path.name}'/'{img_filename}' (rename_on_conflict=False).")
                    overwritten_count += 1

            # --- Process Label File (Change Class IDs using mapping) ---
            label_processed_ok = process_label_file(lbl_source_path, lbl_dest_path, class_id_mapping)

            if label_processed_ok:
                processed_lbl_count += 1
                # Copy corresponding image file to TARGET destination (if it exists and path is set)
                if img_source_path and img_dest_path:
                    try:
                        shutil.copy2(img_source_path, img_dest_path) # copy2 preserves metadata
                        processed_img_count += 1
                    except Exception as e:
                         logging.error(f"    ERROR copying image '{img_source_path.name}' to '{img_dest_path}': {e}")
                         error_count += 1
                         # Optional: Delete label if image copy fails
                         # if lbl_dest_path.exists(): lbl_dest_path.unlink()
                         # processed_lbl_count -= 1
                elif not img_dest_path and img_filename:
                    pass # Image source dir was missing, label processed, image skipped
            else:
                # Label processing failed or no relevant lines found
                error_count += 1 # Count as error/warning


        print(f"  Split '{split}' summary:")
        print(f"    Labels Processed/Written to Target (with mapped ID(s) {list(class_id_mapping.values())}): {processed_lbl_count}")
        print(f"    Images Copied to Target: {processed_img_count}")
        print(f"    Skipped (due to conflicts/existing renamed): {skipped_conflict_count}")
        print(f"    Overwritten (rename_on_conflict=False): {overwritten_count}")
        print(f"    Errors/Warnings during processing: {error_count}")

    print("\n--- Finished: Processing and ADDING NEW Blackline Dataset Files to Target Dataset ---")

else:
    print("--- Skipping Processing: One or both dataset base directories not found. ---")


# --- Verification Steps (Optional but recommended) ---
# You can run the 'Verify Class IDs' script (ID: verify_class_ids) afterwards
# to confirm the class ID range is still 0-20 and the file counts have increased.
print("\n--- Verification Recommended ---")
print("Run the 'Verify Class IDs' script (ID: verify_class_ids) to check the final label integrity.")
print("File counts should be higher, and the class ID range should still be 0-20.")
print("---------------------------------")

```

```python
# --- Modified Removal Script (Ignores Count Mismatch) ---
import os
from pathlib import Path
import logging
import time
import sys

# Setup logging
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')

# --- Configuration ---
# *** Verify these paths are correct ***
source_dataset_dir = Path('/content/drive/MyDrive/BlackLine2') # Used ONLY to get the list of files to potentially remove
target_dataset_dir = Path('/content/drive/MyDrive/Combine_dataset') # The dataset to be cleaned

source_dataset_name = "Blackline2" # For logging
target_dataset_name = "Combine_dataset" # For logging

# Set to False first for a dry run (only lists files to be deleted)
# Set to True to actually delete files after reviewing the dry run output
PERFORM_DELETION = True # <--- START WITH False !!! Run Dry Run First!

splits_to_process = ['train', 'valid', 'test']
image_subdir = 'images'
label_subdir = 'labels'
# Define common image extensions (lowercase)
image_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.webp']
label_ext = '.txt'
# --- End Configuration ---


# --- Helper Function to Get Relative Paths ---
def get_relative_file_paths(base_dir, splits, image_subdir, label_subdir, image_exts, label_ext):
    """ Scans dataset splits/subfolders and returns a set of relative file paths. """
    relative_paths = set()
    # Basic logging inside function for clarity during execution
    # logging.info(f"Scanning '{base_dir.name}' for helper function...")
    if not base_dir.is_dir():
        logging.error(f"Helper Fn: Base directory not found: {base_dir}")
        return relative_paths
    for split in splits:
        split_path = base_dir / split
        if not split_path.is_dir(): continue
        img_path = split_path / image_subdir
        if img_path.is_dir():
            try:
                for item in img_path.iterdir():
                    if item.is_file() and item.suffix.lower() in image_exts:
                        # Use os.path.join for better cross-platform relative paths
                        relative_path_str = os.path.join(split, image_subdir, item.name)
                        relative_paths.add(relative_path_str)
            except Exception as e: logging.error(f"Helper Fn: Error reading {img_path}: {e}")
        lbl_path = split_path / label_subdir
        if lbl_path.is_dir():
            try:
                for item in lbl_path.iterdir():
                    if item.is_file() and item.suffix.lower() == label_ext:
                         # Use os.path.join for better cross-platform relative paths
                        relative_path_str = os.path.join(split, label_subdir, item.name)
                        relative_paths.add(relative_path_str)
            except Exception as e: logging.error(f"Helper Fn: Error reading {lbl_path}: {e}")
    # logging.info(f"Helper Fn: Found {len(relative_paths)} relative paths in '{base_dir.name}'.")
    return relative_paths

# --- Main Removal Script (Modified Strategy) ---
print("--- Remove Source Files from Target Script (Ignoring Count Mismatch) ---")
print(f"SOURCE Dataset (for file list): {source_dataset_dir}")
print(f"TARGET Dataset (to be cleaned): {target_dataset_dir}")
print(f"Identification Method: Exact relative path and filename match.")
if PERFORM_DELETION:
    print(" WARNING: PERFORM_DELETION is TRUE. Files WILL be permanently deleted.")
else:
    print("  INFO: PERFORM_DELETION is FALSE. Running in DRY RUN mode.")
print("--------------------------------------------------------------------")

# --- Basic Path Validation ---
if not source_dataset_dir.is_dir() or not target_dataset_dir.is_dir():
     logging.critical("!!! CRITICAL ERROR: One or both dataset base directories NOT FOUND. Aborting.")
     sys.exit("Required dataset directory not found.")

# --- Step 1: Get files from SOURCE dataset ---
logging.info(f"Step 1: Scanning SOURCE dataset '{source_dataset_name}' to get reference file list...")
source_files_set = get_relative_file_paths(
    source_dataset_dir, splits_to_process, image_subdir, label_subdir, image_extensions, label_ext
)
source_file_count = len(source_files_set)
logging.info(f"Step 1 Result: Found {source_file_count} files in SOURCE dataset '{source_dataset_name}'.")

# Exit if source is empty - cannot determine what to delete
if source_file_count == 0:
    logging.warning("Source dataset appears empty or couldn't be read. Cannot identify files to remove.")
    sys.exit("Source dataset empty or unreadable.")

# --- Step 2: Get files from TARGET dataset ---
logging.info(f"\nStep 2: Scanning TARGET dataset '{target_dataset_name}'...")
target_files_set = get_relative_file_paths(
    target_dataset_dir, splits_to_process, image_subdir, label_subdir, image_extensions, label_ext
)
target_file_count = len(target_files_set)
logging.info(f"Step 2 Result: Found {target_file_count} total files in TARGET.")

# --- Step 3: Identify Matching Files ---
logging.info("\nStep 3: Identifying files present in BOTH Source and Target...")
# Find intersection - files present in BOTH source and target (with same relative path)
matching_relative_paths = source_files_set.intersection(target_files_set)
target_match_count = len(matching_relative_paths) # This is the count of files to potentially delete

# Generate list of full paths in TARGET for the matching files
files_to_remove_full_paths = []
logging.info(f"Constructing full paths for the {target_match_count} matching file(s)...")
for rel_path_str in matching_relative_paths:
    # Construct full path within the TARGET directory using Path object
    # Convert string relative path to Path parts for joining
    full_path_in_target = target_dataset_dir.joinpath(*Path(rel_path_str).parts)
    if full_path_in_target.is_file(): # Double check it exists in target before adding
       files_to_remove_full_paths.append(full_path_in_target)
    else:
       logging.warning(f"Matching relative path '{rel_path_str}' identified, but corresponding file not found at {full_path_in_target} in Target. Skipping removal for this.")

logging.info(f"Step 3 Result: Identified {len(files_to_remove_full_paths)} existing files in TARGET to potentially remove (exact matches from Source).")

# --- Step 4: Report Counts (Informational Only) ---
# Count check is removed as a blocker, but we still log the info
logging.info(f"\nStep 4: Count Summary (Informational):")
logging.info(f"   - Source Files Total: {source_file_count}")
logging.info(f"   - Target Files Total: {target_file_count}")
logging.info(f"   - Matching Files Found in Target: {target_match_count}") # Should equal len(files_to_remove_full_paths) if no warnings above
if source_file_count != target_match_count:
    logging.warning(f"   - NOTE: Source count ({source_file_count}) does NOT match Target match count ({target_match_count}).")
    logging.warning(f"     As requested, script will proceed to remove the {len(files_to_remove_full_paths)} matching files found.")
else:
    logging.info("   - Counts match.")


# --- Step 5: Remove Matching Files ---
print("\n--- Step 5: Removal Execution ---")
files_deleted_count = 0
if not files_to_remove_full_paths:
     logging.info("No matching files were found in the Target dataset to remove.")
elif PERFORM_DELETION:
    print(f" Attempting to DELETE {len(files_to_remove_full_paths)} files from '{target_dataset_name}'...")
    print(f"      Waiting 5 seconds... (Press 'Stop' button in Colab to cancel)")
    time.sleep(5)
    for file_path in files_to_remove_full_paths:
        try:
            file_path.unlink() # Delete the file
            # Use WARNING level for deletions to make them stand out
            logging.warning(f"  DELETED: {file_path}")
            files_deleted_count += 1
        except Exception as e:
            logging.error(f"  ERROR deleting {file_path}: {e}")
    print(f"\nDeletion complete. {files_deleted_count} files removed.")
else:
    # Dry Run Mode
    print("  DRY RUN Mode. Listing files that WOULD be deleted:")
    print_limit = 50 # Limit how many filenames are printed
    if files_to_remove_full_paths:
        for i, file_path in enumerate(files_to_remove_full_paths):
            if i < print_limit:
                 # Use relative path for cleaner logging if possible
                 try:
                     rel_display_path = file_path.relative_to(target_dataset_dir)
                 except ValueError: # Handle cases where it might not be relative (shouldn't happen here)
                     rel_display_path = file_path
                 print(f"  Would delete: {rel_display_path}")
            elif i == print_limit:
                 print(f"  ... (listing truncated, {len(files_to_remove_full_paths) - print_limit} more not shown)")
                 break
        print(f"\nDry run complete. {len(files_to_remove_full_paths)} files identified for potential removal.")
        print("Set PERFORM_DELETION = True and re-run to delete these files.")
    else:
         print("   (No matching files found to list for deletion).")


print("---------------------------------")

# --- End Script ---
```

```python
import os
import shutil
import glob
import logging
from pathlib import Path
import sys # Import sys for exit

# Setup logging
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')

# --- Configuration ---

# 1. Path to the EXISTING dataset FOLDER TO BE MODIFIED
target_dataset_base_dir = Path('/content/drive/MyDrive/Combine_dataset') # <--- VERIFY THIS

# 2. Path to the NEW dataset containing classes to merge
# *** UPDATED to Blackline3 ***
new_dataset_base_dir = Path('/content/drive/MyDrive/BlackLine3') # <--- VERIFY THIS PATH

# 3. Class ID Mapping: Map IDs from NEW dataset (Blackline3) to TARGET dataset (Combine_dataset)
# *** USER ACTION: Please verify this mapping is exactly what you need ***
# *** Especially check the mapping for Source ID 4 ('lanes') ***
class_id_mapping = {
    0: 4,  # Blackline3 'Speed Limit -20-' -> Combine 'Speed Limit -20-'
    1: 5,  # Blackline3 'Speed Limit -30-' -> Combine 'Speed Limit -30-'
    2: 6,  # Blackline3 'Speed Limit -40-' -> Combine 'Speed Limit -40-'
    3: 12, # Blackline3 'Stop Sign'        -> Combine 'Stop Sign'
    4: 20  # Blackline3 'lanes'            -> Combine 'line'  (*** ASSUMPTION: Change 20 to 19 if it should be 'dotted-line' ***)
}
# Expected final number of classes and max ID in Combine_dataset (DO NOT CHANGE unless target dataset changes)
final_nc = 21
final_max_id = 20

# --- Dataset Structure Settings ---
use_images_labels_subfolders_TARGET = True # For Combine_dataset

# --- Filename Conflict Handling ---
rename_new_files_on_conflict = True # Keep True (safer)
# *** Use a new prefix specific to this merge operation ***
rename_prefix = "bl3_" # <--- New prefix for Blackline3 merge

# Define common image extensions (add if needed)
image_extensions = ['.png', '.jpg', '.jpeg', '.bmp', '.webp']
# --- End Configuration ---

# --- Print configuration for verification ---
print("--- Configuration Summary (Adding Blackline3 Data) ---")
print(f"Target (Existing) Dataset Path: {target_dataset_base_dir} <--- WILL BE MODIFIED")
print(f"  Structure uses images/labels subfolders: {use_images_labels_subfolders_TARGET}")
print(f"New Dataset Path (Source): {new_dataset_base_dir}")
print(f"Class ID Mapping (Original Source ID -> Final Target ID):")
# Print mapping clearly
for src_id, trg_id in class_id_mapping.items():
    # You could load YAMLs here to show names, but IDs are crucial
    print(f"  Source Class ID {src_id}  -> Target Class ID {trg_id}")
print(f"Rename new files on conflict: {rename_new_files_on_conflict}")
if rename_new_files_on_conflict:
    print(f"  Rename prefix: '{rename_prefix}'")
print("-----------------------------")
print(" WARNING: The TARGET dataset folder listed above will be directly modified. ")
print(" It is strongly recommended to BACK UP the target folder before proceeding. ")
print("-----------------------------")

# --- Basic Path Validation ---
if not target_dataset_base_dir.is_dir():
    logging.critical(f"!!! CRITICAL ERROR: Target dataset directory NOT FOUND: {target_dataset_base_dir}")
    sys.exit("Target dataset not found.") # Force exit
if not new_dataset_base_dir.is_dir():
    logging.critical(f"!!! CRITICAL ERROR: New dataset directory NOT FOUND: {new_dataset_base_dir}")
    sys.exit("New dataset not found.") # Force exit

# --- Function to process a single label file ---
def process_label_file(lbl_source_path, lbl_dest_path, mapping):
    """Reads a label file, modifies class IDs based on mapping, writes to dest."""
    modified_lines = []
    lines_processed_count = 0
    lines_skipped_unmapped = 0
    lines_skipped_malformed = 0
    try:
        with open(lbl_source_path, 'r') as infile:
            for line_num, line in enumerate(infile, 1):
                parts = line.strip().split()
                if len(parts) >= 5: # Basic check for YOLO format (class + 4 coords)
                    try:
                        original_class_id = int(parts[0])
                        if original_class_id in mapping:
                            # Map the class ID to the target ID
                            parts[0] = str(mapping[original_class_id])
                            modified_lines.append(" ".join(parts))
                            lines_processed_count += 1
                        else:
                            # Log only once per file about unmapped IDs found
                            # logging.debug(f"Skipping line {line_num} in '{lbl_source_path.name}': Original ID {original_class_id} not in mapping keys {list(mapping.keys())}.")
                            lines_skipped_unmapped += 1
                    except ValueError:
                        logging.warning(f"Malformed line (non-int class ID) in '{lbl_source_path.name}', line {line_num}. Skipping: '{line.strip()}'")
                        lines_skipped_malformed += 1
                elif line.strip(): # Line is not empty but doesn't have enough parts
                    logging.warning(f"Malformed line (too few parts) in '{lbl_source_path.name}', line {line_num}. Skipping: '{line.strip()}'")
                    lines_skipped_malformed += 1

        # Write modified label file to TARGET destination only if valid lines were mapped
        if lines_processed_count > 0:
            lbl_dest_path.parent.mkdir(parents=True, exist_ok=True) # Ensure dir exists
            with open(lbl_dest_path, 'w') as outfile:
                outfile.write("\n".join(modified_lines) + "\n")

            log_level = logging.INFO # Default log level for success
            log_message_suffix = ""
            if lines_skipped_unmapped > 0:
                log_message_suffix += f" Skipped {lines_skipped_unmapped} lines (unmapped ID)."
                log_level = logging.INFO # Keep as INFO even if some lines skipped
            if lines_skipped_malformed > 0:
                 log_message_suffix += f" Skipped {lines_skipped_malformed} lines (malformed)."
                 log_level = logging.WARNING # Upgrade to WARNING if malformed lines present

            logging.log(log_level, f"    Processed '{lbl_source_path.name}' -> '{lbl_dest_path.name}': {lines_processed_count} lines mapped.{log_message_suffix}")
            return True # Indicate success (at least some lines mapped)
        else:
            # Log if the file contained only classes we didn't map or was empty/malformed
            logging.info(f"    Skipped writing '{lbl_dest_path.name}': No lines with mappable class IDs {list(mapping.keys())} found in '{lbl_source_path.name}'. (Unmapped: {lines_skipped_unmapped}, Malformed: {lines_skipped_malformed})")
            return False # Indicate no lines relevant to the mapping were processed

    except FileNotFoundError:
         logging.error(f"ERROR reading label file '{lbl_source_path.name}': File not found.")
         return False
    except Exception as e:
        logging.error(f"ERROR processing label file '{lbl_source_path.name}': {e}")
        return False # Indicate error


# --- Main Processing Loop ---
print("\nPlease carefully review the paths, settings, and warnings above.")
print("--- Starting: Processing and ADDING NEW Blackline3 Dataset Files to Target Dataset ---")

total_processed_lbl_count = 0
total_processed_img_count = 0
total_skipped_conflict_count = 0
total_overwritten_count = 0
total_error_count = 0

if target_dataset_base_dir.is_dir() and new_dataset_base_dir.is_dir(): # Proceed only if both base dirs exist
    for split in ['train', 'valid', 'test']:
        print(f"\nProcessing NEW dataset split: {split} -> adding to TARGET '{split}'")

        # Determine NEW dataset source paths (Blackline3)
        new_split_dir = new_dataset_base_dir / split
        new_img_src_dir = new_split_dir / 'images'
        # *** Assuming labels are in 'labels' subdir based on previous script ***
        new_lbl_src_dir = new_split_dir / 'labels'

        # Determine TARGET dataset destination paths (Combine_dataset)
        target_img_dest_dir = target_dataset_base_dir / split / 'images' if use_images_labels_subfolders_TARGET else target_dataset_base_dir / split
        target_lbl_dest_dir = target_dataset_base_dir / split / 'labels' if use_images_labels_subfolders_TARGET else target_dataset_base_dir / split

        processed_img_count_split = 0
        processed_lbl_count_split = 0
        skipped_conflict_count_split = 0
        overwritten_count_split = 0
        error_count_split = 0

        # Ensure target destination directories exist (create if necessary)
        try:
            target_img_dest_dir.mkdir(parents=True, exist_ok=True)
            target_lbl_dest_dir.mkdir(parents=True, exist_ok=True)
        except OSError as e:
            logging.error(f"Could not create target directories for split {split}: {e}. Skipping split.")
            continue

        # Check if source directories exist for the new dataset's split
        if not new_lbl_src_dir.is_dir():
            logging.info(f"  Info: NEW label source directory not found for split '{split}', nothing to add from here: {new_lbl_src_dir}")
            continue # Skip this split if no source labels folder
        if not new_img_src_dir.is_dir():
            logging.warning(f"  Warning: NEW image source directory not found for split '{split}', will process labels but not copy images: {new_img_src_dir}")

        print(f"  Processing labels from '{new_lbl_src_dir}'...")
        # Iterate through label files first
        try:
            label_files_in_new = list(new_lbl_src_dir.glob('*.txt'))
            logging.info(f"  Found {len(label_files_in_new)} .txt files in source label directory.")
        except Exception as e:
            logging.error(f"  Could not list files in source label directory {new_lbl_src_dir}: {e}")
            continue # Skip split if cannot read source labels

        if not label_files_in_new:
             logging.info(f"  No .txt files found in {new_lbl_src_dir}, skipping label processing for split '{split}'.")
             continue

        for lbl_source_path in label_files_in_new:
            base_name = lbl_source_path.stem # Filename without extension
            logging.debug(f"  Checking label file: {lbl_source_path.name}") # Debug level

            # Find corresponding image file in the source image directory
            img_filename = None
            img_source_path = None
            found_corresponding_image = False
            if new_img_src_dir.is_dir(): # Only look for images if src dir exists
                for ext in image_extensions:
                    potential_img_path = new_img_src_dir / (base_name + ext)
                    if potential_img_path.exists():
                        img_filename = potential_img_path.name
                        img_source_path = potential_img_path
                        found_corresponding_image = True
                        logging.debug(f"    Found corresponding image: {img_filename}")
                        break # Stop searching once found

            if not found_corresponding_image and new_img_src_dir.is_dir():
                logging.warning(f"    Warning: No corresponding image found for label '{lbl_source_path.name}' in source image dir '{new_img_src_dir}' using extensions {image_extensions}. Skipping this file pair.")
                error_count_split += 1
                continue # Skip this label file if its image is missing
            elif not new_img_src_dir.is_dir():
                # Allow processing labels even if image dir is missing, image copy will be skipped later
                logging.info(f"    Info: Processing label '{lbl_source_path.name}' without image check (source image dir missing).")
                img_filename = None
                img_source_path = None # Ensure we don't try to copy image later


            # --- Determine Destination Filenames & Handle Conflicts ---
            dest_lbl_filename = lbl_source_path.name
            dest_img_filename = img_filename # Can be None if image dir was missing

            lbl_dest_path = target_lbl_dest_dir / dest_lbl_filename
            img_dest_path = target_img_dest_dir / dest_img_filename if dest_img_filename else None

            # Check for conflicts
            conflict_exists = lbl_dest_path.exists() or (img_dest_path and img_dest_path.exists())
            current_rename_prefix = "" # Use original name unless conflict forces rename

            if conflict_exists:
                if rename_new_files_on_conflict:
                    logging.info(f"    Info: Conflict for '{lbl_source_path.name}' / '{img_filename}'. Renaming with '{rename_prefix}' prefix.")
                    current_rename_prefix = rename_prefix # Set prefix for this file

                    # Create new prefixed names
                    dest_lbl_filename = f"{current_rename_prefix}{lbl_source_path.name}"
                    dest_img_filename = f"{current_rename_prefix}{img_filename}" if img_filename else None

                    # Update destination paths
                    lbl_dest_path = target_lbl_dest_dir / dest_lbl_filename
                    img_dest_path = target_img_dest_dir / dest_img_filename if dest_img_filename else None

                    # Check again if even the renamed file exists
                    if lbl_dest_path.exists() or (img_dest_path and img_dest_path.exists()):
                        logging.warning(f"    Warning: Renamed file '{dest_lbl_filename}' / '{dest_img_filename}' also exists. Skipping this file pair.")
                        skipped_conflict_count_split += 1
                        continue # Skip to next label file
                else: # rename_new_files_on_conflict is False, so we will overwrite
                    logging.warning(f"    Warning: Overwriting existing file(s) for '{lbl_source_path.name}' / '{img_filename}' (rename_on_conflict=False).")
                    overwritten_count_split += 1
                    # Allow processing to continue, will overwrite later

            # --- Process Label File (Change Class IDs using mapping) ---
            label_processed_ok = process_label_file(lbl_source_path, lbl_dest_path, class_id_mapping)

            if label_processed_ok:
                processed_lbl_count_split += 1
                # Copy corresponding image file to TARGET destination (if it exists and path is set)
                if img_source_path and img_dest_path:
                    try:
                        logging.debug(f"    Copying '{img_source_path.name}' to '{img_dest_path.name}'")
                        shutil.copy2(img_source_path, img_dest_path) # copy2 preserves metadata
                        processed_img_count_split += 1
                    except Exception as e:
                        logging.error(f"    ERROR copying image '{img_source_path.name}' to '{img_dest_path}': {e}")
                        error_count_split += 1
                        # If image copy fails, maybe remove the label file we just wrote?
                        logging.warning(f"    Attempting to remove label '{lbl_dest_path.name}' due to image copy error.")
                        try:
                             if lbl_dest_path.exists():
                                 lbl_dest_path.unlink()
                                 processed_lbl_count_split -= 1 # Decrement count as label is removed
                                 logging.info(f"    Removed '{lbl_dest_path.name}'.")
                        except Exception as del_e:
                             logging.error(f"    ERROR removing label '{lbl_dest_path.name}' after image copy failed: {del_e}")

                elif not img_source_path and new_img_src_dir.is_dir():
                    # This case should have been caught earlier by the 'continue'
                    logging.error("    Internal Logic Error: Should not reach image copy stage if image source path wasn't found.")
                    error_count_split += 1
                # If img_source_path is None because new_img_src_dir was missing, do nothing (label processed, no image)

            # else: # label_processed_ok is False
                # This means no mappable lines were found in the label file, or an error occurred during processing.
                # Do not increment error count here, process_label_file already logged info/error.
                # We also don't copy the image if the label wasn't processed successfully.
                # logging.debug(f"Label file {lbl_source_path.name} not processed successfully, image not copied.")


        # Add split counts to totals
        total_processed_lbl_count += processed_lbl_count_split
        total_processed_img_count += processed_img_count_split
        total_skipped_conflict_count += skipped_conflict_count_split
        total_overwritten_count += overwritten_count_split
        total_error_count += error_count_split

        print(f"\n  Split '{split}' summary:")
        print(f"    Labels Processed/Written to Target: {processed_lbl_count_split}")
        print(f"    Images Copied to Target: {processed_img_count_split}")
        print(f"    Skipped (due to conflicts/existing renamed): {skipped_conflict_count_split}")
        print(f"    Overwritten (rename_on_conflict=False): {overwritten_count_split}")
        print(f"    Errors/Warnings during processing (e.g., missing images, copy errors): {error_count_split}")

    print("\n--- Finished: Processing and ADDING NEW Blackline3 Dataset Files to Target Dataset ---")
    print("\n--- Overall Summary ---")
    print(f"  Total Labels Processed/Written: {total_processed_lbl_count}")
    print(f"  Total Images Copied: {total_processed_img_count}")
    print(f"  Total Skipped Conflicts: {total_skipped_conflict_count}")
    print(f"  Total Overwritten: {total_overwritten_count}")
    print(f"  Total Errors/Warnings: {total_error_count}")
    print("-------------------------")


else:
    print("--- Skipping Processing: One or both dataset base directories not found. ---")


# --- Verification Steps (Optional but recommended) ---
print("\n--- Verification Recommended ---")
print("1. Check the logs above for any ERROR or WARNING messages.")
print("2. Manually inspect a few images/labels in the target dataset splits")
print(f"   to confirm files were added (possibly with '{rename_prefix}' prefix) and labels have correct IDs.")
# print("3. Run the 'Verify Class IDs' script (ID: verify_class_ids) afterwards")
# print("   to confirm the class ID range is still 0-20 and file counts have increased.")
print("---------------------------------")
```

```python
import os
import glob

# --- Configuration ---
# Directory containing the dataset to analyze
dataset_dir_to_analyze = '/content/drive/MyDrive/Combine_dataset/'
dry_run = False # Set to False to actually perform modifications (USE WITH CAUTION & BACKUP)
# --- End Configuration ---

if dry_run:
    print("!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")
    print("!!!               DRY RUN MODE ENABLED                   !!!")
    print("!!! NO FILES WILL BE MODIFIED in this run.               !!!")
    print(f"!!! Analyzing Directory: {dataset_dir_to_analyze}")
    print("!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!\n")
else:
    # This part is just for safety - ideally, use the separate modifying script
    print("!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")
    print("!!! WARNING: DRY RUN IS DISABLED - SCRIPT WILL MODIFY FILES !!!")
    print(f"!!! Target Directory: {dataset_dir_to_analyze}")
    print("!!! MAKE SURE YOU HAVE A BACKUP of this directory before proceeding.")
    print("!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")
    # import time
    # time.sleep(5)


print("Starting dataset analysis process...")

# --- Process each split (train, valid, test - add/remove as needed) ---
splits_to_process = ['train', 'valid']
for split in splits_to_process:
    print(f"\n===== Analyzing Split: {split} =====")

    label_dir = os.path.join(dataset_dir_to_analyze, split, 'labels')

    if not os.path.exists(label_dir):
        print(f"WARNING: Label directory not found: {label_dir}. Skipping this split.")
        continue

    # --- Step 1: Analyze Label Format ---
    print(f"\n--- Step 1: Analyzing Label Format ---")
    label_files = glob.glob(os.path.join(label_dir, '*.txt'))
    print(f"Found {len(label_files)} label files to check.")
    format_issues_found_files = [] # List to store filenames with format issues

    for label_file in label_files:
        try:
            lines_would_change = False
            with open(label_file, 'r') as f_in:
                lines = f_in.readlines()

            if not lines: continue # Skip empty files

            for line in lines:
                parts = line.strip().split()
                if len(parts) > 5:  # Format issue detected
                    lines_would_change = True
                    break # No need to check further lines in this file

            if lines_would_change:
                format_issues_found_files.append(os.path.basename(label_file))

        except Exception as e:
            print(f"Error analyzing file {label_file} during format check: {e}")

    print(f"Format Analysis: Found {len(format_issues_found_files)} files that *would* be modified due to format issues (extra coordinates).")
    if format_issues_found_files:
        print("Files with format issues:")
        for filename in format_issues_found_files:
            print(f"  - {filename}")

    # --- Step 2: Analyze for Duplicate Lines ---
    print(f"\n--- Step 2: Analyzing for Duplicate Lines ---")
    # Analyze the same set of files again for duplicates
    # In a real run, we'd analyze the potentially format-fixed files, but for dry run, analyze original
    label_files_to_check_duplicates = glob.glob(os.path.join(label_dir, '*.txt'))
    print(f"Found {len(label_files_to_check_duplicates)} label files to check for duplicates.")
    duplicates_found_files = [] # List to store filenames with duplicates

    for label_file in label_files_to_check_duplicates:
        try:
            lines_read = []
            with open(label_file, 'r') as f_in:
                lines_read = [line.strip() for line in f_in if line.strip()]

            if not lines_read: continue # Skip empty files

            # Use set to find unique lines count
            unique_line_count = len(set(lines_read))

            # Check if duplicates exist
            if unique_line_count < len(lines_read):
                duplicates_found_files.append(os.path.basename(label_file))

        except Exception as e:
            print(f"Error analyzing file {label_file} during duplicate check: {e}")

    print(f"Duplicate Analysis: Found {len(duplicates_found_files)} files containing duplicate lines that *would* be removed.")
    if duplicates_found_files:
        print("Files with duplicate lines:")
        for filename in duplicates_found_files:
            print(f"  - {filename}")


print("\n===== Dataset Analysis (Dry Run) Finished =====")
if dry_run:
    print("NO FILES WERE MODIFIED.")
```

```python
import os
import glob
import time # For safety delay

# --- Configuration ---
# *** This script modifies files IN PLACE in this directory ***
dataset_dir_to_fix = '/content/drive/MyDrive/Combine_dataset/'
# --- End Configuration ---

# --- !!! BACKUP WARNING !!! ---
print("!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")
print("!!!      WARNING: THIS SCRIPT WILL MODIFY FILES DIRECTLY    !!!")
print(f"!!!      Target Directory: {dataset_dir_to_fix}")
print("!!!      MAKE SURE YOU HAVE A BACKUP before proceeding.     !!!")
print("!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")
# Adding a small delay to allow reading the warning
print("Starting modification in 5 seconds...")
time.sleep(5)


print("\nStarting dataset fixing process (in-place modification)...")

# --- Process each split (train, valid, test - add/remove as needed) ---
splits_to_process = ['train', 'valid']
for split in splits_to_process:
    print(f"\n===== Processing Split: {split} =====")

    label_dir = os.path.join(dataset_dir_to_fix, split, 'labels')

    if not os.path.exists(label_dir):
        print(f"WARNING: Label directory not found: {label_dir}. Skipping this split.")
        continue

    # --- Step 1: Fix Label Format (In-Place) ---
    print(f"\n--- Step 1: Fixing Label Format (In-Place) ---")
    label_files_step1 = glob.glob(os.path.join(label_dir, '*.txt'))
    print(f"Found {len(label_files_step1)} label files to check for format.")
    format_modified_count = 0

    for label_file in label_files_step1:
        try:
            lines_changed = False
            original_lines = []
            corrected_lines = []

            # Read original content first
            with open(label_file, 'r') as f_in:
                original_lines = f_in.readlines()

            if not original_lines: continue # Skip empty files

            # Process lines
            for line in original_lines:
                parts = line.strip().split()
                if len(parts) > 5:  # Format issue detected
                    corrected_lines.append(" ".join(parts[:5]) + "\n") # Keep only first 5 parts
                    lines_changed = True
                elif len(parts) >= 0: # Keep valid or potentially empty lines
                     corrected_lines.append(line)

            # If changes were made, overwrite the original file
            if lines_changed:
                format_modified_count += 1
                with open(label_file, 'w') as f_out:
                    f_out.writelines(corrected_lines)

        except Exception as e:
            print(f"Error processing file {label_file} during format fix: {e}")

    print(f"Format Fix: Modified format for {format_modified_count} files.")

    # --- Step 2: Remove Duplicate Lines (In-Place) ---
    print(f"\n--- Step 2: Removing Duplicate Lines (In-Place) ---")
    label_files_step2 = glob.glob(os.path.join(label_dir, '*.txt'))
    print(f"Found {len(label_files_step2)} label files to check for duplicates.")
    duplicates_removed_count = 0

    for label_file in label_files_step2:
        try:
            lines_read = []
            # Read the current content (potentially format-fixed)
            with open(label_file, 'r') as f_in:
                lines_read = [line.strip() for line in f_in if line.strip()] # Read non-empty lines

            if not lines_read: continue # Skip empty files

            # Use list to preserve order while checking uniqueness
            ordered_unique_lines = []
            seen_lines = set()
            for line in lines_read:
                if line not in seen_lines:
                    ordered_unique_lines.append(line)
                    seen_lines.add(line)

            # Check if duplicates were found and removed
            if len(ordered_unique_lines) < len(lines_read):
                duplicates_removed_count += 1
                # Overwrite the file with unique lines
                with open(label_file, 'w') as f_out:
                    for unique_line in ordered_unique_lines:
                        f_out.write(unique_line + "\n")

        except Exception as e:
            print(f"Error processing file {label_file} during duplicate removal: {e}")

    print(f"Duplicate removal: Removed duplicates from {duplicates_removed_count} files.")
    print(f"Files in {label_dir} should now be fixed in-place.")

print("\n===== Dataset Fixing Process (In-Place Modification) Finished =====")
print("Double-check your files and verify missing labels manually.")
print("Remember to delete the .cache files from the dataset directory before training.")
```
